```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
## dealing problems with outlier removing!!
```

```python
def remove_outliers(df_1, col_value):

    df_1_cols = df_1.columns.tolist()
    df_2 = df_1[df_1_cols[:col_value]]
    df_3 = df_1[df_1_cols[col_value:]]
    df_3_cols = df_1_cols[col_value:]

    low = .01
    high = .99
    quantile_df = df_3.quantile([low, high]) 

    for ele_i in df_3_cols:
        low_bound = quantile_df[ele_i].iloc[0]
        high_bound = quantile_df[ele_i].iloc[1]
        df_3.loc[df_3[ele_i] <= low_bound, ele_i] = np.NaN
        df_3.loc[df_3[ele_i] >= high_bound, ele_i] = np.NaN
        
    df_4 = pd.concat([df_2, df_3], axis=1)
    return df_4
```

```python
eicu_1 = pd.read_csv("eicu_raw_data.csv")
mimic3_1 = pd.read_csv("mimic3_raw_data.csv")
mimic4_1 = pd.read_csv("mimic4_raw_data.csv")
display(eicu_1)
display(mimic3_1)
display(mimic4_1)
```

```python
m4_cols = mimic4_1.columns.tolist()
e_cols = eicu_1.columns.tolist()
m3_cols = mimic3_1.columns.tolist()

print(np.array(m3_cols))
print(np.array(m4_cols))
print(np.array(e_cols))
```

```python
eicu_2 = remove_outliers(eicu_1, 7)
mimic3_2 = remove_outliers(mimic3_1, 7)
mimic4_2 = remove_outliers(mimic4_1, 19)
```

```python
e_c1 = np.array(eicu_2.columns.tolist())
m3_c1 = np.array(mimic3_2.columns.tolist())
m4_c1 = np.array(mimic4_2.columns.tolist())
print(e_c1)
print(m3_c1)
print(m4_c1)
```

```python
eicu_3 = eicu_2.copy()
eicu_3 = eicu_3.rename(columns={"SUBJECT_ID":"hadm_id",
                                "GENDER":"gender",
                                "AGE":"age",
                                "AKI_label":"had_AKI", 
                                "creatinine_charttime":"AKI_start_time", 
                                "lab_time":"charttime"})
print(eicu_3.shape[0])

mimic3_3 = mimic3_2.copy()
mimic3_3 = mimic3_3.rename(columns={"HADM_ID":"hadm_id", 
                                    "GENDER":"gender", 
                                    "AGE":"age", 
                                    "CREATININE_CHARTTIME":"AKI_start_time", 
                                    "AKI_label":"had_AKI", 
                                    "CHARTTIME":"charttime"})
print(mimic3_3.shape[0])

mimic4_3 = mimic4_2.copy()
mimic4_3 = mimic4_3.rename(columns={"AKI_label":"had_AKI", 
                                    "creatinine_charttime":"AKI_start_time"})
print(mimic4_3.shape[0])
```
