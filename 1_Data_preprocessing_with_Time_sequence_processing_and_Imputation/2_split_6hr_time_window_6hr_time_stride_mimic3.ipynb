```python
import pandas as pd
import numpy as np
```

```python
pd1 = pd.read_csv("1_cleaned_mimic3.csv")
print(pd1.shape[0])
pd1.head()
```

```python
# select patients whose age >= 18
pd1 = pd1[pd1['age'] >= 18]
pd1.shape[0]
```

```python
data = pd1.copy()

#change datatype of AKI label
data["had_AKI"] = data["had_AKI"].astype("bool")

#cast both datetime columns to datetime
data["charttime"] = pd.to_datetime(data["charttime"])
data["AKI_start_time"] = pd.to_datetime(data["AKI_start_time"])

#replace with a bool for easier analysis
data["is_male"] = data["gender"] == "M"

#I think these are already sorted but it doesn't hurt
data = data.sort_values(["hadm_id", "charttime"])

#fix ordering
data = data[["hadm_id", "charttime", "had_AKI", "AKI_start_time",
             "is_male", "age"] + list(data.columns[6:-1])]
data.head()
```

```python
data
```

```python
#window_size = string, how many hours in a sliding window (ex. "2 hours")
#stride_size = how many hours the sliding window slides by (ex. "2 hours")
#fill_na_with_last_value = fills subsequent missing values with the previous value for that reading
#fill_na_with_last_value = fills any remaining NA values with the mean value for that reading
#keep_post_aki_data defaults to false, this throws out any data after the window with AKI 
#(does not throw out the window with the AKI)
def make_windows(data, window_size, window_stride, 
                 fill_na_with_last_value = False,
                 fill_remaining_na_with_mean = False,
                 keep_post_aki_data = False):
    
    encounter_rows = []
    hadm_id_group = data.groupby('hadm_id')
    hadm_id_group_size_count = hadm_id_group.size().reset_index(
        name='counts')
    hadm_id_array = hadm_id_group_size_count['hadm_id'].to_numpy()
    df_append_ls = []
    window_size = pd.to_timedelta(window_size)
    window_stride = pd.to_timedelta(window_stride)
    
    for i, hid in enumerate(hadm_id_array):
        one_p = len(hadm_id_array)//100       
        print("Processed", i+1, "of" ,len(hadm_id_array)+1)
        
        rows = []
        encounter = hadm_id_group.get_group(hadm_id_array[i])
        hadm_id = encounter.iloc[0, 0]
        encounter_start_time = encounter.iloc[0, 1]  
        had_aki = encounter.iloc[0, 2]
        aki_time = encounter.iloc[0, 3]
        is_male = encounter.iloc[0, 4]
        age = encounter.iloc[0, 5]
        last_event_time = encounter.iloc[-1]["charttime"]
        sample_row = encounter.iloc[0]
        
        window_start = encounter_start_time
        
        while window_start <= last_event_time:
            points_in_window = encounter[(window_start <= encounter["charttime"]) & (encounter["charttime"] < window_start + window_size)]
            aki_in_window = (window_start <= aki_time) & (aki_time < window_start + window_size)
            
            
            row = pd.concat([pd.Series({"HADM_ID":hadm_id, 
                                        "WINDOW_START":window_start, 
                                        "WINDOW_END":window_start + window_size, 
                                        "AKI_IN_WINDOW":aki_in_window, 
                                        "AKI_IN_ENCOUNTER":had_aki, 
                                        "AGE":age, "IS_MALE":is_male}), points_in_window.iloc[:,6:].mean(axis=0)])
            if len(rows) > 0 and fill_na_with_last_value:
                row = row.fillna(rows[-1])
            
            rows.append(row)
            window_start += window_stride
            
            if aki_in_window and not keep_post_aki_data:
                break
        encounter_rows.append(pd.concat(rows, axis=1).T)
    fixed_data =  pd.concat(encounter_rows, axis=0)
    
    if fill_remaining_na_with_mean:
        fixed_data = fixed_data.iloc[:,7:].fillna(data.iloc[:,6:].mean())
        
    return fixed_data
```

```python
#you can write this to a CSV file if you want
#change the window size and stride before you run this to whatever you need it to be
#smaller values for window size will take longer to run
prepared_data = make_windows(data, window_size="6 hours", window_stride ="6 hours", 
                             fill_na_with_last_value=True, 
                             fill_remaining_na_with_mean=False)
```

```python
#this counting the non-NA values in 2 day windows 
#i.e about 1 row per encounter in this version of the dataset. 
#Many of these variables are rarely seen in patients.
(~prepared_data.isna()).mean().sort_values(ascending=False)
```
