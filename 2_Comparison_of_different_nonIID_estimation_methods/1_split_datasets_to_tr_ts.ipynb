```python
path2 = '/home/ai03/_kunyi/pro_shae/fed_aki_9/compute_non_iid_1_compare_different_methods/predict_6hr_tr_ts_datasets/'
```

```python
def get_datasets(m_df, dy):
    
    m3_df2 = m_df.sample(
        frac=1).reset_index(drop=True)
    m3_arr_5f = np.array_split(m3_df2, 5)
    
    m3_arr_5f_1 = m3_arr_5f[0]
    m3_arr_5f_2 = m3_arr_5f[1]
    m3_arr_5f_3 = m3_arr_5f[2]
    m3_arr_5f_4 = m3_arr_5f[3]
    m3_arr_5f_5 = m3_arr_5f[4]

    m3_arr_5f_2345 = pd.concat([m3_arr_5f_2, m3_arr_5f_3, 
                                m3_arr_5f_4, m3_arr_5f_5]) 
    m3_arr_5f_2345 = m3_arr_5f_2345.sample(
        frac=1).reset_index(drop=True)

    m3_arr_5f_1345 = pd.concat([m3_arr_5f_1, m3_arr_5f_3, 
                                m3_arr_5f_4, m3_arr_5f_5]) 
    m3_arr_5f_1345 = m3_arr_5f_1345.sample(
        frac=1).reset_index(drop=True)

    m3_arr_5f_1245 = pd.concat([m3_arr_5f_1, m3_arr_5f_2, 
                                m3_arr_5f_4, m3_arr_5f_5]) 
    m3_arr_5f_1245 = m3_arr_5f_1245.sample(
        frac=1).reset_index(drop=True)

    m3_arr_5f_1235 = pd.concat([m3_arr_5f_1, m3_arr_5f_2, 
                                m3_arr_5f_3, m3_arr_5f_5]) 
    m3_arr_5f_1235 = m3_arr_5f_1235.sample(
        frac=1).reset_index(drop=True)
    
    m3_arr_5f_1234 = pd.concat([m3_arr_5f_1, m3_arr_5f_2, 
                                m3_arr_5f_3, m3_arr_5f_4]) 
    m3_arr_5f_1234 = m3_arr_5f_1234.sample(
        frac=1).reset_index(drop=True)
    
    m3_arr_5f_1.to_csv(path2 + "{}_ts_1.csv".format(dy),index=False)
    m3_arr_5f_2.to_csv(path2 + "{}_ts_2.csv".format(dy),index=False)
    m3_arr_5f_3.to_csv(path2 + "{}_ts_3.csv".format(dy),index=False)
    m3_arr_5f_4.to_csv(path2 + "{}_ts_4.csv".format(dy),index=False)
    m3_arr_5f_5.to_csv(path2 + "{}_ts_5.csv".format(dy),index=False)
    
    m3_arr_5f_2345.to_csv(path2 + "{}_tr_1.csv".format(dy),index=False)
    m3_arr_5f_1345.to_csv(path2 + "{}_tr_2.csv".format(dy),index=False)
    m3_arr_5f_1245.to_csv(path2 + "{}_tr_3.csv".format(dy),index=False)
    m3_arr_5f_1235.to_csv(path2 + "{}_tr_4.csv".format(dy),index=False)
    m3_arr_5f_1234.to_csv(path2 + "{}_tr_5.csv".format(dy),index=False)
```

```python
get_datasets(m3_df, 'm3')
get_datasets(m4_df, 'm4')
get_datasets(e_df, 'e_0.6')
```
